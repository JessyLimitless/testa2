```python
import matplotlib.pyplot as plt

# Sample function to create a stock price graph
def make_graph(stock_data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data['Date'], stock_data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Stock Price ($)")
    plt.legend()
    plt.grid(True)
    plt.show()

# Sample Tesla data for demonstration
tesla_stock_data = pd.DataFrame({
    "Date": ["2023-10-20", "2023-10-19", "2023-10-18", "2023-10-17", "2023-10-16",
             "2023-10-13", "2023-10-12", "2023-10-11", "2023-10-10", "2023-10-09"],
    "Close": [250.13, 245.28, 247.39, 249.52, 243.80, 240.45, 242.32, 244.00, 243.25, 241.50]
})

# Create and display the Tesla stock graph
make_graph(tesla_stock_data, "Tesla Stock Price Over Time")
```

```python
# Sample GameStop data for demonstration
gme_stock_data = pd.DataFrame({
    "Date": ["2023-10-20", "2023-10-19", "2023-10-18", "2023-10-17", "2023-10-16",
             "2023-10-13", "2023-10-12", "2023-10-11", "2023-10-10", "2023-10-09"],
    "Close": [17.72, 17.45, 17.21, 17.69, 17.52, 17.32, 17.15, 17.48, 17.50, 17.20]
})

# Create and display the GameStop stock graph
make_graph(gme_stock_data, "GameStop Stock Price Over Time")
```
